## Digits Dataset

La base de datos Digits contiene imágenes de dígitos escritos a mano (del 0 al 9).

* Tiene 1,797 muestras en total
* Cada imagen es de 8×8 píxeles
* Los valores de los píxeles van de 0 a 16 (niveles de gris discretos)
* Incluye etiquetas correspondientes al dígito representado

Se usa principalmente para:
* Ejemplos rápidos y pruebas en aprendizaje automático
* Enseñanza de clasificación sin necesidad de grandes recursos

En esencia, Digits es como una versión “ligera” de MNIST: menos datos, menor resolución, pero ideal para aprender y prototipar modelos sin preocuparse por tiempos de entrenamiento largos.

```python
print(digits.DESCR)
```

## Visualización de una imagen

### Obtener la imagen del dígito y su etiqueta

```python
index = 0
image = digits.images[index]
label = digits.target[index]
```

### Imprimimos la imagen

```python
print(image)
```

### Mostramos la imagen con Matplotlib

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(3, 3))
plt.imshow(image, cmap=plt.cm.gray)
plt.title(f'Dígito: {label}')
plt.show()
```
