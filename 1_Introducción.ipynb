### Modelos de aprendizaje

Los modelos de aprendizaje automático se dividen en aprendizaje supervisado y no supervisado (principalmente).

### Aprendizaje supervisado

En el aprendizaje supervisado, los modelos se entrenan utilizando datos etiquetados.

Esto significa que cada ejemplo de entrada (por ejemplo, una imagen) viene acompañado de una salida correcta (como la etiqueta “gato” o “perro”).

El objetivo del modelo es aprender que relacione correctamente las entradas con sus salidas, minimizando el error entre sus predicciones y las etiquetas reales.

### CNNs

Las redes convolucionales (CNNs) surgieron a finales de los años 80's y principios de los 90's, siendo pioneros en este campo

* Kunihiko Fukushima en su trabajo *Neocognitron: A Self-organizing Neural Network Model for a Mechanism of Pattern Recognition Unaffected by Shift in Position en 1980*
* Yann LeCun en su trabajo *Backpropagation Applied to Handwritten Zip Code Recognition en 1989.*

LeCun desarrolló una arquitectura conocida como LeNet-5, que fue una de las primeras CNN utilizadas para tareas de reconocimiento de dígitos escritos a mano. https://www.youtube.com/watch?v=FwFduRA_L6Q&t=8s

Las CNNs son una forma de redes neuronales artificiales en las cuales las neuronas tratan de imitar a las neuronas en la corteza visual primaria de un cerebro biológico.  

También, las podemos considerar como una variante de los perceptrones multicapa, solo que las entradas de esta red son matrices bidimensionales (o tensores), la cual las hace altamente efectivas para aplicaciones de visión artificial, como la clasificación y segmentación de imágenes, entre otras tareas.

### Convolución matricial

La operación de convolución matricial entre una matriz de entrada $X$ y un filtro $W$ se define de la siguiente manera:
Dada una matriz de entrada $X$ de tamaño $n \times m$ y un filtro $W$ de tamaño $p\times q$ con $n>p$ y $m>q$, al mapa de características resultante  se le conoce como convolución y se denota como:
$$X \odot W= Z$$

Donde  

* $Z_{i,j}$ es el elemento en la posición del mapa de características $$Z_{i,j} = \sum_{u=1}^{p} \sum_{v=1}^{q} X_{i+u-1, j+v-1} \cdot W_{u,v} $$

* $X_{i+u-1, j+v-1}$ son los elementos de la matriz de entrada afectados por el filtro en la posición $u$, $v$.

* $W_{u,v}$ son los pesos del filtro en la posición $u$, $v$ .

### Redes Convolucionales

Una CNN es un tipo de modelo de aprendizaje profundo diseñado específicamente para el procesamiento de datos estructurados de manera jerárquica, como imágenes o secuencias.

La definición de una CNN se puede desglosar en varios componentes clave:

### Pre-procesamiento

La red toma como entrada los pixeles de una imagen.

image.png

### Capas de Convolución

Las capas de convolución son la característica central de las CNNs. Utilizan filtros (también llamdos kernels) para realizar operaciones de convolución sobre la entrada.


Cada filtro detecta patrones específicos en la entrada y genera mapas de características que resaltan esas características.

image.png

### Funciones de Activación

Después de la convolución, se aplica una función de activación no lineal, (normalmente se utiliza la función ReLU), para introducir caracteristicas no lineales en el modelo.  

Esto permite a la red aprender relaciones más complejas en los datos.

image.png

### Capas de Agrupamiento (Pooling)

Las capas de agrupamiento reducen la dimensionalidad de los mapas de características generados por las capas de convolución.  

Esto se logra mediante operaciones como tomar el máximo o el promedio de un conjunto de valores (matriz), lo que ayuda a conservar la información importante mientras se reduce el tamaño de los datos.

image.png

### Capas Totalmente Conectadas

Después de varias capas de convolución y agrupamiento, se pueden agregar capas totalmente conectadas para realizar tareas específicas, como clasificacar o realizar regresión.
